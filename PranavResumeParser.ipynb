```python
from pdfminer.high_level import extract_text
def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(extract_text_from_pdf("resumetest.pdf"))
doc_cleaned = [token for token in doc if not token.is_stop and not token.is_punct]
```

```python
# Extract Email
list_of_Emails=[]
for entity in doc:
    if entity.like_email:
        list_of_Emails.append(entity)
print(list_of_Emails)
```

```python
# Extract Language
list_of_LANGUAGE=[]

for entity in doc.ents:
  if entity.label_=='LANGUAGE':
    list_of_LANGUAGE.append(entity.text)
print(list_of_LANGUAGE)
```

```python
# Extract URLs
list_of_URLs=[]
for entity in doc:
    if entity.like_url:
        list_of_URLs.append(str(entity))
print(list_of_URLs)
dictionary_of_URLs={}
otherurllist=[]
for i in list_of_URLs:
    if('linkedin' in i):
        dictionary_of_URLs['linkedin']=i
    elif('github' in i):
        dictionary_of_URLs['github']=i
    elif('youtube' in i):
        dictionary_of_URLs['youtube']=i
    elif('facebook' in i):
        dictionary_of_URLs['facebook']=i
    else:
        otherurllist.append(i)
if len(otherurllist)!=0:
    dictionary_of_URLs['other']=otherurllist
print(dictionary_of_URLs)
```
